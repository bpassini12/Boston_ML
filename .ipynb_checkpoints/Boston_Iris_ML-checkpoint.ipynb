```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

# Boston Linear Example

```python
boston = load_boston()
X,y = boston.data, boston.target
n_samples, n_features = X.shape
print(n_samples, n_features)
```

```python
print(boston.feature_names)
```

```python
plt.hist(y)
plt.xlabel('Price', fontsize=18)
```

```python
df = pd.DataFrame(X, columns=boston.feature_names)
df.head()
```

### The intercept (Theta_0)

```python
model = LinearRegression()
model.fit(X, y)
print(model.intercept_) 
```

### The coefficients (theta_1, ..., ...theta_5)

```python
print(model.coef_.shape)
```

```python
model.fit(X[::2], y[::2])
print('R2 score: %s' % model.score(X[1::2], y[1::2]))
```

```python
from sklearn import linear_model
dir(linear_model)
```

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha = 0.1)
model.fit(X, y)
print(model.intercept_)    #the intercept (theta_0)
```

```python
print(model.coef_.shape)   # the coefficients (theta_1, ..., theta_5)
```

# Linear Classification

```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
iris = datasets.load_iris()
X = iris.data[:, :2]  # make 2D
y = iris.target

X, y = X[y<2], y[y<2] # make it binary

y[y==0] = -1
print(X.shape)
```

```python
print(np.unique(y))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=1.)
model.fit(X,y)
theta0 = model.intercept_  # the intercept (theta0)
theta = model.coef_[0]     # the coefficients (theta1, ..., theta5)
```

## Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
model = SVC(kernel= 'linear', C = 1.)
model.fit(X, y)
theta0 = model.intercept_
theta = model.coef_[0]
```
